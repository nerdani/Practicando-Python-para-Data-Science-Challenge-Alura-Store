```python
import matplotlib.pyplot as plt

# Contar productos por categoría
conteo_categorias = {}
for categoria in tienda["Categoría del Producto"]:
    conteo_categorias[categoria] = conteo_categorias.get(categoria, 0) + 1

# Ordenar de mayor a menor
categorias_ordenadas = sorted(conteo_categorias, key=conteo_categorias.get, reverse=True)
cantidades_ordenadas = [conteo_categorias[c] for c in categorias_ordenadas]

# Gráfico de barras
plt.bar(categorias_ordenadas, cantidades_ordenadas, color='skyblue')
plt.title("Cantidad de productos por categoría")
plt.xlabel("Categoría")
plt.ylabel("Cantidad")
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas
plt.tight_layout()  # Ajustar espacio para etiquetas
plt.show()
```
